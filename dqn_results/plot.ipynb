```python
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
import optax
epsilon_by_frame = optax.polynomial_schedule(init_value=1.0, end_value=0.05, 
							                     transition_steps=500, power=0.999)
```

```python
folders = [e for e in os.listdir('.') if os.path.isdir(os.path.join('.', e))]
folders.sort()
for folder in folders:
    if not os.path.exists(f'{folder}/reward.csv'):
        continue
    rewards = np.loadtxt(f'{folder}/reward.csv', delimiter=',')
    losses = np.loadtxt(f'{folder}/loss.csv', delimiter=',')
    fig, axs = plt.subplots(1, 2, figsize=(12, 3), dpi=200)
    fig.suptitle(folder)
    axs[0].set_title('Reward')
    axs[0].set_xlabel('Episode')
    axs[0].set_ylabel('Reward')
    axs[0].plot(rewards)
    axs[1].set_title('Loss')
    axs[1].set_xlabel('Episode')
    axs[1].set_ylabel('Loss')
    axs[1].plot(losses)
    fig.tight_layout()
    plt.show()
```
